# The Hundred 2023 Best 11

I will create My 11 for The Hundred 2023 backed by data and personal choices.

Criteria: 
Atleast 5 innings batted or bowled.
Atleast 100 runs for Batsman and 6 wickets for Bowler.

Playing 11 Composition
1. Opener with Excellent Average and Strike Rate
2. Opener with High Strike Rate
3. Batsman with Excellent Average and Strike Rate
4. Batsman with High Boundary Count and Strike Rate
5. Batsman with High Strike Rate
6. All Rounder with High Strike Rate and some Wickets
7. All Rounder with Good Strike Rate and good Wickets
8. Spinner with Highest Wickets and Less Economy
9. Bowler with high wickets in Death over
10. Bowler with high wickets in Powerplay as well as Middle and Death
11. Bowler with high PowerPlay Wickets

Number 1-5 should have WicketKeeper and Captain

Number 1, someone who can play steady innings with High Average because ball has swung the most.

Number 2, Opener hitter to combat the swing of the ball.

Number 3,4,5 can be flexible.

Number 3, someone who can play long innings with good Strike Rate while their partner takes on Bowler.

Number 4, someone who has high boundary count and ability to hit sixes to spinner

Number 5, someone who has good Strike Rates

At Number 6,7 AR who has high strike and some wicket taking ability is great asset in shorter version of game. Bowling at Death over is another good to have.

At Number 8 Spinner who has ability to take wicket in Middle Phase.

At Number 9 Faster who has ability to take wicket in Death Phase.

At Number 10 Faster who has ability to take wicket in PowerPlay and Middle(Death) Phase.

At Number 11 Faster who has ability to take wicket in PowerPlay Phase.

More emphasis on PowerPlay wicket taker because, Ball has swung the most in English Tournament in The Hundred than any other.

I think bowling is vital in shorter version and as well known mantra is "Batting wins you a Game and Bowling wins you a Tournament"

Reference: https://www.espncricinfo.com/story/the-hundred-2023-stats-analysis-why-opening-is-harder-1393524
